# Belgrave Valley Chess Tournament

Welcome to the BV Chess Tournament pairing organizer! This script is designed to facilitate the pairing process for the upcoming BV Chess Tournament. The tournament will be held over two separate days, with six rounds each day and three matches per round.

The purpose of this script is to efficiently pair participants for each round, ensuring fair and balanced matchups throughout the tournament. By automating the pairing process, we aim to streamline the organization of the tournament and provide a seamless experience for both participants and organizers.

With clear and concise pairings generated for each round, players can focus their energy on the game itself, rather than the logistics of scheduling matches.

So let's get started and make the first BV Chess Tournament a memorable and successful event for all involved!

## Library importation and participants

```python
import pandas as pd
import numpy as np
import random


Dimid= ['Cuts','Chobo','Claudio','NacNac','Javas','Enri']
Pakin= ['Juan','Braulio','Pablo','Paco','Manu','Marcos']
```

```python
# @title Random Matchup Generator
def GameGenerator2(player1,player2,gamesperround, roundperday):
  Sat_games={}
  Sun_games={}
  dimi=player1.copy()
  paki=player2.copy()
  possible_games = []
  normal_games = []
  Nacho_games=[]
  for i in range(roundperday):
    Sat_games[i+1]=[]
    Sun_games[i+1]=[]

  #Mezclamos aleatoriamente el orden de equipos
  random.shuffle(dimi)
  random.shuffle(paki)

  df = pd.DataFrame(data=dimi, columns=['T1'])
  df2 = pd.DataFrame(data=paki, columns=['T2'])
  df['join'] = 1
  df2['join'] = 1

  df_output = pd.merge(df,df2, on='join')


  for i in range(len(df_output['join'])):
    possible_games.append(df_output.loc[i]['T1'] + " vs " +df_output.loc[i]['T2'] )


  for i in range(len(possible_games)):
    if 'NacNac' in possible_games[i]:
       Nacho_games.append(possible_games[i])
    else:
      normal_games.append(possible_games[i])


  for i in range(len(Sat_games)):
    while len(Sat_games[i+1]) < gamesperround and normal_games:
      random.shuffle(normal_games)
      game = normal_games.pop()
      if all(game.split(' vs ')[0] != Sat_games[i + 1][j].split(' vs ')[0] and game.split(' vs ')[0] != Sat_games[i + 1][j].split(' vs ')[1] and game.split(' vs ')[1] != Sat_games[i + 1][j].split(' vs ')[0] and game.split(' vs ')[1] != Sat_games[i + 1][j].split(' vs ')[1] for j in range(len(Sat_games[i + 1]))):
        Sat_games[i + 1].append(game)
      else:
            normal_games.append(game)

  #Todas las partidas de nacho se asignan al domingo
  for i in range(len(Nacho_games)):
    Sun_games[i+1].append(Nacho_games[i])


  for i in range(len(Sun_games)):
    while len(Sun_games[i+1]) < gamesperround and normal_games:
      random.shuffle(normal_games)

      game = normal_games.pop()
      if all(game.split(' vs ')[0] != Sun_games[i + 1][j].split(' vs ')[0] and game.split(' vs ')[0] != Sun_games[i + 1][j].split(' vs ')[1] and game.split(' vs ')[1] != Sun_games[i + 1][j].split(' vs ')[0] and game.split(' vs ')[1] != Sun_games[i + 1][j].split(' vs ')[1] for j in range(len(Sun_games[i + 1]))):
        Sun_games[i + 1].append(game)
      else:
        normal_games.append(game)

  new_key_names = {'Round 1': 1, 'Round 2': 2, 'Round 3': 3, 'Round 4': 4, 'Round 5': 5, 'Round 6': 6}

  Output_saturday = {}
  Output_sunday = {}
  for new_key, old_key in new_key_names.items():
    Output_saturday[new_key] = Sat_games[old_key]
    Output_sunday[new_key] = Sun_games[old_key]


  df_sat=pd.DataFrame.from_dict(Output_saturday).transpose( ).reset_index(names="Round")
  df_sun=pd.DataFrame.from_dict(Output_sunday).transpose( ).reset_index(names="Round")
  df_sat['Day']="Saturday"
  df_sun['Day']="Sunday"
  df_games= pd.concat([df_sat,df_sun],axis=0)
  df_games

  return(df_games)
```

```python
# @title Color Assigning formula (Black & White)
def ColorAssign(Matchups, team1,team2,mixes):
  random.shuffle(team1)
  random.shuffle(team2)

  players_team1_code= {}
  players_team2_code= {}
  for i in range(len(team1)):
    players_team1_code["A"+str(i+1)] = team1[i]
    players_team2_code["B"+str(i+1)] = team2[i]

  totalplayers= players_team1_code

  totalplayers.update(players_team2_code)





  for i in range(len(mixes)):
    # Loop through each key-value pair in the dictionary
    for key, value in totalplayers.items():
        # Check if the key is in the current item
        if key in mixes[i]:
            # Replace the key with the corresponding value
            mixes[i] = mixes[i].replace(key, value)

  mixes.sort()

  players = list(totalplayers.values())
  count_players = []
  played_with_white={}
  played_with_black=[]

  for i in mixes:
    count_players.append(i.split(", ",1)[1].replace(' plays white',""))

  for i in mixes:
    if ' plays black' in i:
      played_with_black.append(i.split(',')[0].strip())

  for i in played_with_black:
    count_players.append(i.split('vs')[1].strip())



  for i in players:
    played_with_white[i]=count_players.count(i)


  final = pd.DataFrame(mixes, columns=["color_assign"])
  return final

```
